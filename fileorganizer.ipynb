```python
import zipfile
import random
import os
import gradio as gr
import time
import shutil
```

```python
folname={"documents":[".txt",".pdf",".doc",".docx",".odt",".rtf",".md"],
"spreadsheet":[".xls",".xlsx",".csv","tsv",".ods"],
"presentation":[".ppt",".pptx",".odp",".key"],
"image":[".jpg",".jpeg",".png",".gif",".webp",".bmp",".svg"],
"audio":[".mp3",".wav",".aac",".ogg",".flac"],
"video":[".mp4",".mkv",".avi",".mov",".webm"],
"compressed":[".zip",".rar",".7z",".tar",".gz"],
"program":[".py",".java",".c",".cpp",".js",".html",".css",".php"],
"executable":[".exe",".apk",".bat",".sh",".msi"],
"web":[".json",".xml",".yml",".php"],
"database":[".db",".sqlite",".sql",".mdb"],
"system":[".ini",".cfg",".log",".env"],
"email":[".eml",".msg",".pst"]}




folname

```

```python
def find_key(ext):
    for k,v in folname.items():
        if (ext in v) and k != "" and v != "" :
            return k
    return "others"
```

```python
def seperate(files):
    print("files")
    dirs=[]
    data=os.listdir(files)
    
    print(data)
    for file in data:
        ext=os.path.splitext(file)[1]
        fol=find_key(ext)
        
        if fol not in dirs:
            print("----")
            os.makedirs(os.path.join(files, fol), exist_ok=True)
            print("----")
            
            dirs.append(fol)
        os.path.exists(f"{files}\\{file}")
        os.path.exists(f"{files}\\{fol}\\{file}")
        shutil.move(f"{files}\\{file}",f"{files}\\{fol}\\{file}")
    print("returned")
       
        
```

```python

def fileclassifer(file):
    name=str(time.time())
    os.mkdir(name)
    with zipfile.ZipFile(file,"r") as zip_ref:
         zip_ref.extractall(name)
    print("extracted")
    seperate(name)
    
    shutil.make_archive(base_name=f"outputs\\{name}\\{name}",format="zip",root_dir=f"D:\\projects\\file_seprater\\{name}")
    print("done")
    return f"outputs\\{name}\\{name}.zip"

    
```

```python

theme=gr.themes.Soft(
    primary_hue="blue",
    neutral_hue="slate"
)
with gr.Blocks(theme=theme) as app:
    files_zip=gr.File(file_count="single",file_types=[".zip"],label="upload zip files")

    
    btn=gr.Button("classify the file")
    output=gr.File(label="download")
    btn.click(fileclassifer,files_zip,output)

```
